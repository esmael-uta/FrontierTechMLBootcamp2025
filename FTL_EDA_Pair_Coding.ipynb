**Step 3: Basic Data Overview**


```python
# Check dataset structure
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

**Step 4: Handling Missing Data**

```python
# Fill missing numeric values with median
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns  # Select numeric columns only
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Check if missing values are handled
print(df.isnull().sum())
```

**Step 5: Data Visualization**

***A. Histogram for Distribution of Features***

```python
df.hist(figsize=(12, 8), bins=30)
plt.show()
```

***B. Correlation Heatmap***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

# Add title
plt.title("Correlation Heatmap of California Housing Features")
plt.show()

```

***C. Scatter Plot of Median House Value vs. Median Income***

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['median_income'], y=df['median_house_value'])
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Income vs. House Value')
plt.show()
```

***D.  Box Plot for House Value by Ocean Proximity***

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['ocean_proximity'], y=df['median_house_value'])
plt.xticks(rotation=45)
plt.show()
```

***E.  Pairplot for Key Features***

```python
sns.pairplot(df[['median_income', 'total_rooms', 'housing_median_age', 'median_house_value']])
plt.show()
```
